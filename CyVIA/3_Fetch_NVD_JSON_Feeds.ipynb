```python
# NVD JSON Handler...

# Imports
import os, glob, requests, re, zipfile, timeit
from os import listdir
from datetime import datetime
from time import time as timer
from os.path import isfile, join

# global variables and other classes
file_path = 'nvd_json_files' # JSON files path globally
nvd_feeds_url = 'https://nvd.nist.gov/vuln/data-feeds#JSON_FEED'
feed_urls = []

import functions as fn # functions class
func = fn.functions()

# Check available feed files and generate URLs
def check_available_files(feeds_url):
    # Start obtaining new feeds...
    print('Checking NVD feeds from: '+feeds_url)
    r = requests.get(feeds_url) # where the files are...
    for filename in re.findall('nvdcve-1.1-[0-9]*\.json\.zip', r.text): # yearly file names e.g. nvdcve-1.1-2021.json.zip 
        feed_urls.append('https://nvd.nist.gov/feeds/json/cve/1.1/'+filename)
    
# Download files
def fetch_url_file(url):
    # Start obtaining new feeds...
    print('Obtaining '+url+'...')
    r = requests.get(url) # where the files are...
    r_file = requests.get(url, stream=True)
    with open(file_path + '/' + os.path.basename(url), 'wb') as f: # path, file name from URL
        for chunk in r_file.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)

    
print('*** NIST\'s NVD JSON Handler v4 ***\n')
# time the execution
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Start Time:", current_time)
startTime = datetime.now() # start timer

# Step 1: Check and remove existing files and folders
func.check_files(file_path)

# Step 2: Check available feeds
check_available_files(nvd_feeds_url)

# Step 3: Download feed files
start = timer()
from multiprocessing.pool import ThreadPool
results = ThreadPool(8).imap_unordered(fetch_url_file, feed_urls)
for path in results:
    print('Done') # print(path)

# Step 3: extract and remove zip files
func.extract_and_remove_zips(file_path)

print('\nData collected.\nFile(s) saved at '+file_path+'/.')
print("Execution time: "+str(datetime.now() - startTime))
```
